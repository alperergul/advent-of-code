```python
with open("test.txt", "r") as file:
    data = file.read()
```

```python
len(data)
```

```python
dataArray = data.split('\n\n')
dataArray
```

```python
totals = []
for i in dataArray:
    elfCalories = i.split('\n')
    a = 0;
    for n in elfCalories:
        a += int(n);
    totals.append(a)            
```

```python
totals
```

```python
max = 0
for i in totals:
    if( i > max):
        max = i
max
```
